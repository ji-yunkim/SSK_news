```python
import codecs
from collections import OrderedDict


dic = {}

r = codecs.open("article(dokdo).txt","r","UTF-8")

content = ""

for a in r:
    a=a.replace("(","")
    a=a.replace("[","")
    a=a.replace(")","")
    a=a.replace("]","")
    a=a.replace("-"," ")
    a=a.replace(",","")
    content = content+(a.split("\t")[0]+a.split("\t")[2])

lst = content.split(" ")

for i in lst:
    try:
        dic[i] +=1
    except:
        dic[i] =1

sorted_by_value = OrderedDict(sorted(dic.items(), key=lambda x: x[1]))

for q,w in sorted_by_value.items():
    print(q+":"+str(w))
```
