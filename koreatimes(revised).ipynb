```python
# setting

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import urllib.request
import re
import codecs
import os
import html
import glob
from newspaper import Article

#-*- coding: utf-8 -*-
output_path = "urls.csv"

f = codecs.open(output_path,"r","UTF-8")
url_lists_with_newline = f.readlines()

urls_list = []
urls_to_article = []
dates = []
reporters = []
titles = []
articles = []
sitenames = []
times=[]
id = []

for url_element_with_newline in url_lists_with_newline:
    urls_list.append(url_element_with_newline.strip())

driver = webdriver.Chrome('/Users/user/Desktop/chromedriver')
```

```python
numbering=1
for url in urls_list:
    url_without_http = "".join(url.split("http://")[1:])
    if "www.koreatimes.com" in url_without_http:
        form = ""
        title="-"
        content="-"
        driver.get(url)
        dateraw="-"
        article_url = urllib.request.urlopen(url).read()
        soup=BeautifulSoup(article_url, "lxml")
        body = driver.find_element_by_id("print_arti")
        body = body.text
        body = body.replace("\n"," ")
        content = Article(url, language='ko')
        content.download()
        content.parse()
        try:
            t = content.publish_date
            date = t.strftime("%Y-%m-%d")
        except:
            date = driver.find_element_by_class_name("upload_date")
            date = date.text
            date = date.split(" ")[0]
        
        try:
            t = content.publish_date
            time = t.strftime("%H:%M:%S")
        except:
            time = "No meta time given"
          
        try:
            author = driver.find_element_by_class_name("writer")
            author = author.text
            authorlst = author.split(" ")
            if len(authorlst)==2:
                author = authorlst[0]
            elif len(authorlst)==3:
                author = authorlst[1]
        except:
            author = "No meta author given"
        
        try:
            body = body.split(author)[0]
        except:
            print(author)
            print(url)
        text = content.text
        title = content.title
        if len(title)==0:
            try:
                title = driver.find_element_by_id("tit_arti")
                title = title.text
            except:
                print("no title!")
                print("("+title+")")
                print(url)
                pass
        media = "코리아타임즈"
        # list 담기
        dates.append(date)
        times.append(time)
        reporters.append(author)
        articles.append(body)
        sitenames.append(media)
        titles.append(title)
        id.append(numbering)
        sample = str(numbering)+"\t"+media+"\t"+date+"\t"+time+"\t"+author+"\t"+title+"\t"+body
        #print(sample)
        print(numbering)
        numbering = numbering+1
```

```python
rows = zip(id, sitenames, dates, times, reporters, titles, articles)
```

```python
path = "koreatimes.tsv"

f = codecs.open(path,"w","UTF-8")
for row in rows:
    f.write("".join(str(s)+"\t" for s in row)+ "\n")
```
